```python
import numpy as np
from sklearn.datasets import make_regression
```

```python
X, y = make_regression(n_samples= 4, n_features = 1, n_informative = 2, n_targets = 1, noise = 50)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

lr.fit(X,y)
```

```python
print(lr.coef_, lr.intercept_)
```

```python
plt.scatter(X, y)
plt.plot(X, lr.predict(X), color='red')
```
