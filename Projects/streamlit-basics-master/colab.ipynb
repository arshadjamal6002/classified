```python
# Mulit index study for this
#Top Startups -> year wise -> Overall
result_df = pd.DataFrame()

# Iterate over each group by 'year'
for group, data in df.groupby('year'):
    # Group by 'startup' and sum the 'amount', then sort by amount in descending order
    grouped_data = data.groupby('startup')['amount'].sum().sort_values(ascending=False).head()
    
    # Assign the group ('year') as the index for each iteration
    grouped_data = grouped_data.to_frame().T  # Convert Series to DataFrame and transpose it
    
    # Set the 'year' as the index for the result_df
    grouped_data['year'] = group  # Add the 'year' as a column
    
    # Append the grouped data to the result DataFrame
    result_df = pd.concat([result_df, grouped_data])

# Set the 'year' column as index for the final DataFrame
result_df.set_index('year', inplace=True)

# Optionally, you can reset the column order if needed
result_df = result_df.sort_index(axis=1, ascending=True)  # Sort columns if necessary

# Show the result
# st.write(result_df)    
result_df
```
